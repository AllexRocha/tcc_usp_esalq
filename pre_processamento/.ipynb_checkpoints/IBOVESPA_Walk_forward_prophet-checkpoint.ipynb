```python
import pandas as pd

def ler_csv(caminho_arquivo):
    return pd.read_csv(caminho_arquivo, sep=',',usecols=[0,1], decimal=',', header=0)

df = ler_csv('dados/ibovespa.csv')
```

```python
df.rename(columns={'Data':'data','Último': 'y'}, inplace=True)
df['data'] = pd.to_datetime(df['data'], format='%d.%m.%Y')
df['ds'] = df['data']
df.set_index('data', inplace=True)
df = df.resample('W').mean()
df.sort_index(ascending=True, inplace=True)
df['y'] = df['y'].astype('float64')
#df.index = pd.DatetimeIndex(df.index.values, freq=df.index.inferred_freq)
```

```python
df['y'].plot()
```

```python
#train test split

to_row = int(len(df) *0.9)
df_train = df[0:to_row]
df_test = df[to_row:]
```

```python
from prophet import Prophet
import pandas as pd

# Suponha que 'df' seja seu DataFrame e que ele tem duas colunas: 'ds' e 'y'
# 'ds' é a coluna de data e 'y' é a coluna de valor que você deseja prever

# Crie uma instância do Prophet e ajuste o modelo aos seus dados
model = Prophet()
model.fit(df_train)

# Crie um DataFrame para as datas futuras para as quais você deseja fazer previsões
future = model.make_future_dataframe(periods=len(df_test), freq='W')

# Use o modelo para fazer previsões
forecast = model.predict(future)

# 'forecast' é um DataFrame que contém as previsões
print(forecast[['ds', 'yhat']].tail())
```

```python
from prophet import Prophet
import pandas as pd

# Suponha que 'df' seja seu DataFrame e que ele tem duas colunas: 'ds' e 'y'
# 'ds' é a coluna de data e 'y' é a coluna de valor que você deseja prever

# Defina o número de passos para a frente
steps = 1

# Crie uma lista para armazenar as previsões
predictions = []
history = df_train

# Loop sobre o DataFrame em incrementos de 'steps'
for i in range(steps, len(df_test)):
    # Ajuste o modelo aos dados até o passo atual
    model = Prophet()
    model.fit(history)
    
    # Crie um DataFrame para a data futura para a qual você deseja fazer uma previsão
    future = model.make_future_dataframe(periods=steps)
    
    # Use o modelo para fazer a previsão
    forecast = model.predict(future)
    
    # Adicione a previsão à lista de previsões
    predictions.append(forecast.iloc[-1]['yhat'])

# 'predictions' agora contém suas previsões de walk-forward
```

```python
forecast_prev = forecast[len(df_train):]
len(forecast_prev)
```

```python
forecast_prev = forecast_prev[['ds','yhat']]
```

```python
forecast_prev.index = df_test.index
```

```python
forecast_prev.drop('ds', axis=1,inplace=True)
df_test.drop('ds', axis=1,inplace=True)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,4))
plt.plot(df_test, label="Atual")
plt.plot(forecast_prev, label="Previsão")
plt.title('Ibovespa', fontsize=20)
plt.ylabel('Indice', fontsize=16)
plt.legend()
```
